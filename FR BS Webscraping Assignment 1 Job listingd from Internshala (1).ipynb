```python
#Lets extract all the tag having the job titles
titles=soup.find_all('div',class_="heading_4_5 profile")                    
titles
```

```python
#Now we will extractthe text fromthese tags one by one by looping
job_titles=[]
for i in titles:
       job_titles.append(i.text)
job_titles
```

```python
#Do cleaning to remove extra data
job_titles=[]
for i in titles:
       job_titles.append(i.text.replace('\n',' '))
job_titles
```

```python
len(job_titles)
```

```python
#Lets extract all the tag having the company names
companies=soup.find_all('a',class_="link_display_like_text")
companies
```

```python
#Now we will extract the text from these above tags one by one by looping over these tags
company_names=[]
for i in companies:
    company_names.append(i.text.strip())
company_names
```

```python
#Checking the length
len(company_names)
```

```python
#Now lets extract all the tags having the company job ctc
ctc=soup.find_all('div',class_="item_body")  
ctc
```

```python
#Now we will extrct the text from these tags one by one by looping over these tags
jobs_ctc=[]
for i in ctc:
    jobs_ctc.append(i.text)
jobs_ctc
```

```python
#Do cleaning by removing extra data
jobs_ctc=[]
for i in ctc:
    jobs_ctc.append(i.text.replace('\n',' '))
jobs_ctc
```

```python
#Do cleaning by removing extra data by using range function
ctc=[]
for i in range(1,len(jobs_ctc), 3):
    ctc.append(jobs_ctc[i])
ctc
```

```python
len (ctc)
```

```python
#Now lets extract all the tags having the company job apply date
applydate=soup.find_all('div',class_="item_body")  
applydate
```

```python
#Now we will extrct the text from these tags one by one by looping over these tags
jobs_applydate=[]
for i in applydate:
    jobs_applydate.append(i.text)
jobs_applydate
```

```python
#Do cleaning by removing extra data
jobs_applydate=[]
for i in applydate:
    jobs_applydate.append(i.text.replace('\n',' '))
jobs_applydate
```

```python
#Do cleaning by removing extra data by using range function
applydate=[]
for i in range(2,len(jobs_applydate), 3):
    applydate.append(jobs_applydate[i])
applydate
```
