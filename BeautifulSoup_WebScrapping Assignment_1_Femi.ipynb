```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Checked if we have the permission to scrap the webpage.
# We sent get request to the webpage server to get the source code of the page which is to scraped using requests library.

page= requests.get("https://en.wikipedia.org/wiki/Main_Page")
print(page)
```

Got response 200 which means it is permitted to extract data from the website

```python
#To see the content present in the webpage
page.content
```

```python
# Now we will parse this source code by using BeatifulSoup and parser used here is html.parser as source code is in HTML.
soup= BeautifulSoup(page.content)
soup
```

```python
# Prettify will maintain all the indententions here
print(soup.prettify())
```

```python
# Displaying all the header tags.
titles = soup.find_all(['h1', 'h2', 'h3'])
header_tags= []
for i in titles:
    header_tags.append(i.text.replace("\n"," "))

print("The header tags of WIKIPEDIA MAIN PAGE:\n")
header_tags
```
