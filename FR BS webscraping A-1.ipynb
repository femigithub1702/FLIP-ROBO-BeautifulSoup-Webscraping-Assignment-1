Q4) A python program to scrap book name, author name, genre and book review of any 5 books from
‘www.bookpage.com’


Book name
Author name
Genre
Book review of any 5 books from ‘www.bookpage.com’.

IMPORTING LIBRARIES

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
```

```python
# Checking if we have the permission to scrap the webpage.

page= requests.get("https://bookpage.com/reviews")
page
```

```python
# Now we will parse this source code by using BeatifulSoup and parser used here is html.parser as source code is in HTML.
soup=BeautifulSoup(page.text, 'html.parser')
```

```python
# Page title
soup.title.text
```

```python
# Required data lists
book_name =[]
author_name =[]
genre = []
book_review =[]
```

```python
# Now lets extract all html tags with movie Names.
book_data= soup.findAll('div', class_ ='bp-block article-info')
book_data
```

```python
# Collecting column data
for i in book_data:
    name = i.h4.a.text
    book_name.append(name)
    
    author= i.p.text.replace("\n", " ")
    author_name.append(author)
    
    book_genre= i.find('p', class_ = 'genre-links hidden-phone').text.replace("\n"," ")
    genre.append(book_genre)
    
    review = i.find('p', class_='excerpt').text.replace("\n", " ")
    book_review.append(review)
```

```python
# Tabulating data
book_details =  pd.DataFrame({})
book_details['book_name']= book_name
book_details['author_name']= author_name
book_details['genre'] = genre
book_details['book_review'] = book_review
```
